```python
new_data=pd.read_csv('/content/preprocessed_data.csv')
```

```python
new_data.head()
```

```python
new_data.columns
```

```python
from sklearn.preprocessing import StandardScaler
X = new_data.drop(columns=['ANNUAL'])
y = new_data['ANNUAL']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
X_scaled = np.c_[np.ones(X_scaled.shape[0]), X_scaled]
weights = np.zeros(X_scaled.shape[1])
learning_rate = 0.01
n_iterations = 500
for epoch in range(1, n_iterations + 1):
    y_pred = X_scaled @ weights
    error = y_pred - y
    gradient = 2 * X_scaled.T @ error / len(y)
    weights -= learning_rate * gradient
    loss = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    print(f"Epoch {epoch:3}: Loss = {loss:.4f}, R = {r2:.4f}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
X_train_bias = np.c_[np.ones(X_train.shape[0]), X_train]
X_test_bias = np.c_[np.ones(X_test.shape[0]), X_test]
weights = np.zeros(X_train_bias.shape[1])
learning_rate = 0.01
n_iterations = 500
train_r2_list = []
test_r2_list = []
for epoch in range(1, n_iterations + 1):
    y_pred_train = X_train_bias @ weights
    error = y_pred_train - y_train
    gradient = 2 * X_train_bias.T @ error / len(y_train)
    weights -= learning_rate * gradient
    train_r2 = r2_score(y_train, y_pred_train)
    y_pred_test = X_test_bias @ weights
    test_r2 = r2_score(y_test, y_pred_test)
    train_r2_list.append(train_r2)
    test_r2_list.append(test_r2)
    if epoch % 50 == 0 or epoch == 1:
        print(f"Epoch {epoch:3}: Train R = {train_r2:.4f}, Test R = {test_r2:.4f}")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print("Sklearn LinearRegression R (Train):", model.score(X_train, y_train))
print("Sklearn LinearRegression R (Test) :", model.score(X_test, y_test))
```

```python
y_test_pred = X_test_bias @ weights
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_test_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Predicted vs Actual (Custom Linear Regression)")
plt.grid(True)
plt.show()
```
